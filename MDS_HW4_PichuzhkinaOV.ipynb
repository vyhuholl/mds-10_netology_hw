$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1.



Производная первого порядка:
<br>
$f^{\prime}(x)=5x^4+8\cos{(2x)}-3\sin{(3x + 3)}$
<br>
Производная второго порядка:
<br>
$f^{\prime\prime}(x)=20x^3-16\sin{(2x)}-9\cos{(3x+3)}$
<br>
Значения обеих производных в точке x=1:

```
x = 1
```

```
5 * np.power(x, 4) + 8 * np.cos(2 * x) - 3 * np.sin(3 * x + 3)
```

```
20 * np.power(x, 3) - 16 * np.sin(2 * x) - 9 * np.cos(3 * x + 3)
```

# Уровень 1
Постройте график первой и второй производной функции:

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

Производная первого порядка:
<br>
$f^{\prime}(x)=10\sin^{4}{(2x+1)}\cos{(2x+1)}$
<br>
Производная второго порядка:
<br>
$f^{\prime\prime}(x)=80\sin^{3}{(2x+1)}\cos^2{(2x+1)}-20\sin^5{(2x + 1)}$

```
def func_derivative_1(x):
    return 10 * np.power(np.sin(2 * x + 1), 4) * np.cos(2 * x + 1)
```

```
def func_derivative_2(x):
    return 80 * np.power(
        np.sin(2 * x + 1), 3) * np.power(
            np.cos(2 * x + 1), 2) - 20 * np.power(np.sin(2 * x + 1), 5)
```

```
x = np.linspace(-5.0, 5.0, 100)
y_1 = [func_derivative_1(elem) for elem in x]
y_2 = [func_derivative_2(elem) for elem in x]
```

```
fig, (ax_1, ax_2) = plt.subplots(1, 2, sharey=True, figsize=(10, 5))
ax_1.plot(x, y_1)
ax_2.plot(x, y_2)
ax_1.grid(True)
ax_2.grid(True)
ax_1.set_xlabel('X', fontsize='large')
ax_2.set_xlabel('X', fontsize='large')
ax_1.set_ylabel('Y', fontsize='large')
ax_1.set_title('График первой производной f(x)', fontsize='x-large')
ax_2.set_title('График второй производной f(x)', fontsize='x-large')
plt.show()
```

В целом, заметно, что обе функции – это какая-то комбинация из синусов и косинусов, причём во второй коэффициенты и степени больше. (:
